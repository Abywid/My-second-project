```python
import yfinance as yf
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Mengambil data saham (contoh: Google)
stock_data = yf.download('BBCA.jk', start='2019-01-01', end='2025-02-17')

# Memisahkan fitur (X) dan target untuk Open dan Close
X = stock_data[['Open', 'High', 'Low', 'Volume']]  # Fitur yang digunakan
y_open = stock_data['Open']  # Target untuk memprediksi Open
y_close = stock_data['Close']  # Target untuk memprediksi Close

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train_open, y_test_open = train_test_split(X, y_open, test_size=0.2, shuffle=False)
X_train, X_test, y_train_close, y_test_close = train_test_split(X, y_close, test_size=0.2, shuffle=False)

# Membuat dan melatih model Linear Regression untuk Open
model_open = LinearRegression()
model_open.fit(X_train, y_train_open)

# Membuat dan melatih model Linear Regression untuk Close
model_close = LinearRegression()
model_close.fit(X_train, y_train_close)

# Mengambil data hari terakhir dari dataset untuk prediksi besok
last_day_data = stock_data.iloc[-1][['Open', 'High', 'Low', 'Volume']].values.reshape(1, -1)

# Memprediksi harga Open dan Close untuk hari berikutnya (besok)
predicted_open = model_open.predict(last_day_data)
predicted_close = model_close.predict(last_day_data)

# Menampilkan hasil prediksi harga Open dan Close untuk besok
print(f'\nPrediksi harga Open untuk besok: {predicted_open[0][0]:.2f}')
print(f'\nPrediksi harga Close untuk besok: {predicted_close[0][0]:.2f}')
```
